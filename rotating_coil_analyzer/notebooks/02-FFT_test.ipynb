```python
from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from rotating_coil_analyzer.ingest.discovery import MeasurementDiscovery
from rotating_coil_analyzer.ingest.readers_mba import MbaRawMeasurementReader, MbaReaderConfig
from rotating_coil_analyzer.ingest.readers_sm18 import Sm18CorrSigsReader, Sm18ReaderConfig

from rotating_coil_analyzer.models.frames import SegmentFrame
from rotating_coil_analyzer.analysis.turns import split_into_turns
from rotating_coil_analyzer.analysis.fourier import dft_per_turn


# ===========================
# USER SETTINGS
# ===========================
ROOT = Path(r"C:\Local data\2025_12_19-analyzer_tests\867_Alberto\20251212_171026_SPS_MBA\20251212_171620_MBA")
SEGMENT = "CS"         # "CS" or "NCS"
N_MAX = 20             # max harmonic order
REQUIRE_FINITE_T = True  # if True, drop turns with any non-finite t samples (your plateau_id=66 case)


# ===========================
# Helper: enforce full-turn length (explicit statement + trimming)
# ===========================
def ensure_full_turns(segf: SegmentFrame) -> SegmentFrame:
    Ns = int(segf.samples_per_turn)
    n = len(segf.df)
    rem = n % Ns
    if rem != 0:
        n_keep = n - rem
        print(f"INCOMPLETE TURN DETECTED: total samples={n}, Ns={Ns}, remainder={rem}.")
        print(f"Trimming last {rem} samples -> keeping {n_keep} samples before FFT.")
        df2 = segf.df.iloc[:n_keep, :].reset_index(drop=True)
        n_turns2 = n_keep // Ns
        return SegmentFrame(
            source_path=segf.source_path,
            run_id=segf.run_id,
            segment=segf.segment,
            samples_per_turn=Ns,
            n_turns=n_turns2,
            df=df2,
            warnings=tuple(segf.warnings) + (f"analysis trim: removed tail remainder={rem} samples",),
            aperture_id=segf.aperture_id,
        )
    else:
        print(f"Full-turn length OK: total samples={n} is exactly n_turns*Ns ({segf.n_turns}*{Ns}).")
        return segf


# ===========================
# 1) Build catalog and select segment file
# ===========================
cat = MeasurementDiscovery(strict=True).build_catalog(ROOT)

run_id = cat.runs[0]
ap_ui = cat.logical_apertures[0]

seg_id = next(
    s.segment_id for s in cat.segments_for_aperture(ap_ui)
    if str(s.segment_id).upper() == SEGMENT.upper()
)
f = cat.get_segment_file(run_id, ap_ui, seg_id)
print("Using file:", f)

# ===========================
# 2) Read SegmentFrame
# ===========================
if f.name.lower().endswith("_raw_measurement_data.txt"):
    segf = MbaRawMeasurementReader(MbaReaderConfig()).read(
        f,
        run_id=run_id,
        segment=str(seg_id),
        samples_per_turn=cat.samples_per_turn,
        aperture_id=cat.resolve_aperture(ap_ui),
    )
else:
    segf = Sm18CorrSigsReader(Sm18ReaderConfig(strict_time=True)).read(
        f,
        run_id=run_id,
        segment=str(seg_id),
        samples_per_turn=cat.samples_per_turn,
        shaft_speed_rpm=cat.shaft_speed_rpm,
        aperture_id=cat.resolve_aperture(ap_ui),
    )

print("n_turns:", segf.n_turns, "Ns:", segf.samples_per_turn, "n_samples:", len(segf.df))

# ===========================
# 3) Explicit full-turn check (and trim if needed)
# ===========================
segf = ensure_full_turns(segf)

# ===========================
# 4) Split into turns (rows = turns)
# ===========================
tb = split_into_turns(segf)  # tb.t, tb.df_abs, tb.df_cmp, tb.I are (n_turns, Ns)

Ns = int(segf.samples_per_turn)
assert tb.df_abs.shape[1] == Ns and tb.df_cmp.shape[1] == Ns and tb.t.shape[1] == Ns, "Turn reshape mismatch."

# ===========================
# 5) Turn QC: drop bad turns BEFORE FFT, with explicit reporting
# ===========================
ok_abs = np.all(np.isfinite(tb.df_abs), axis=1)
ok_cmp = np.all(np.isfinite(tb.df_cmp), axis=1)

if REQUIRE_FINITE_T:
    ok_t = np.all(np.isfinite(tb.t), axis=1)
else:
    ok_t = np.ones(tb.df_abs.shape[0], dtype=bool)

valid_turn = ok_t & ok_abs & ok_cmp

# --- After computing valid_turn ---
bad_turn_idx = np.where(~valid_turn)[0]
print("\nDROPPED TURN DETAILS")
print("bad turn indices (turn-space):", bad_turn_idx.tolist())

if bad_turn_idx.size:
    # Turn index within its plateau (0..n_turns_in_plateau-1)
    if getattr(tb, "plateau_id", None) is not None:
        pid_turn = tb.plateau_id
        # compute within-plateau turn number
        within = np.zeros_like(pid_turn, dtype=int)
        for pid in np.unique(pid_turn):
            m = np.where(pid_turn == pid)[0]
            within[m] = np.arange(m.size, dtype=int)

        # print a compact table
        rows = []
        for j in bad_turn_idx:
            rows.append({
                "turn_idx": int(j),
                "plateau_id": float(tb.plateau_id[j]) if getattr(tb, "plateau_id", None) is not None else None,
                "plateau_step": float(tb.plateau_step[j]) if getattr(tb, "plateau_step", None) is not None else None,
                "plateau_I_hint": float(tb.plateau_I_hint[j]) if getattr(tb, "plateau_I_hint", None) is not None else None,
                "turn_in_plateau": int(within[j]),
                "bad_t": bool(~ok_t[j]),
                "bad_abs": bool(~ok_abs[j]),
                "bad_cmp": bool(~ok_cmp[j]),
            })
        display(pd.DataFrame(rows))

n_total = int(tb.df_abs.shape[0])
n_drop_t = int(np.sum(~ok_t))
n_drop_abs = int(np.sum(ok_t & ~ok_abs))
n_drop_cmp = int(np.sum(ok_t & ok_abs & ~ok_cmp))
n_keep = int(np.sum(valid_turn))

print("\nTURN QC REPORT")
print("  total turns:", n_total)
print("  keep turns :", n_keep)
print("  drop turns :", n_total - n_keep)
print("   - drop (non-finite t)      :", n_drop_t)
print("   - drop (finite t, bad abs) :", n_drop_abs)
print("   - drop (finite t+abs, bad cmp):", n_drop_cmp)

# If MBA, identify where dropped turns occur
if getattr(tb, "plateau_id", None) is not None:
    bad = np.where(~valid_turn)[0]
    if bad.size:
        uniq, cnt = np.unique(tb.plateau_id[bad], return_counts=True)
        print("  dropped turns by plateau_id:", dict(zip(uniq.tolist(), cnt.tolist())))

# Apply mask
t_turns = tb.t[valid_turn, :]
abs_turns = tb.df_abs[valid_turn, :]
cmp_turns = tb.df_cmp[valid_turn, :]
I_turns = tb.I[valid_turn, :]

# ===========================
# 6) FFT per FDI channel (df_abs and df_cmp), per turn (rows = turns)
# ===========================
H_abs = dft_per_turn(abs_turns, n_max=N_MAX)   # HarmonicsPerTurn(orders, coeff)
H_cmp = dft_per_turn(cmp_turns, n_max=N_MAX)

turn_idx = np.arange(H_abs.coeff.shape[0], dtype=int)

def coeff_table(H, prefix):
    C = H.coeff
    orders = H.orders
    cols = {}
    for j, n in enumerate(orders):
        cols[f"{prefix}A{n}"] = np.abs(C[:, j])
        cols[f"{prefix}phi{n}"] = np.angle(C[:, j])
    return pd.DataFrame(cols, index=turn_idx)

df_h = pd.concat([coeff_table(H_abs, "abs_"), coeff_table(H_cmp, "cmp_")], axis=1)

# ===========================
# 7) Add per-turn metadata (MBA) + mean current, and display results
# ===========================
meta = {}
# Always add I_mean (useful for both SM18 and MBA)
meta["I_mean"] = np.mean(I_turns, axis=1)

# MBA-only metadata (present if split_into_turns exposed it)
if getattr(tb, "plateau_id", None) is not None:
    meta["plateau_id"] = tb.plateau_id[valid_turn]
if getattr(tb, "plateau_step", None) is not None:
    meta["plateau_step"] = tb.plateau_step[valid_turn]
if getattr(tb, "plateau_I_hint", None) is not None:
    meta["plateau_I_hint"] = tb.plateau_I_hint[valid_turn]

df_out = pd.concat([pd.DataFrame(meta, index=turn_idx), df_h], axis=1)

print("\nFirst 10 turns (metadata + a few harmonics):")
cols_preview = []
for n in range(1, min(6, N_MAX + 1)):
    cols_preview += [f"cmp_A{n}", f"cmp_phi{n}"]
preview = [c for c in ["plateau_id", "plateau_step", "plateau_I_hint", "I_mean"] if c in df_out.columns] + cols_preview
display(df_out[preview].head(10))

# ===========================
# 8) Quick diagnostic plot: harmonic amplitudes vs turn index
# ===========================
plt.figure()
if "cmp_A1" in df_out.columns: plt.plot(df_out.index, df_out["cmp_A1"], label="cmp A1")
if "cmp_A2" in df_out.columns: plt.plot(df_out.index, df_out["cmp_A2"], label="cmp A2")
if "cmp_A3" in df_out.columns: plt.plot(df_out.index, df_out["cmp_A3"], label="cmp A3")
plt.xlabel("turn index (after QC)")
plt.ylabel("amplitude |C_n|")
plt.legend()
plt.title(f"Per-turn FFT amplitudes (cmp channel), SEGMENT={SEGMENT}")
plt.show()

# ===========================
# 9) Optional: per-plateau averaging (MBA)
# ===========================
if "plateau_id" in df_out.columns:
    grp = df_out.groupby("plateau_id", sort=True)
    df_plateau = grp.mean(numeric_only=True)
    print("\nPer-plateau mean (first 10 plateaus):")
    display(df_plateau.head(10))

    pid = 116.0
m = (tb.plateau_id == pid)
cmp_turns_pid = tb.df_cmp[m, :]     # turns x Ns within plateau
t_turns_pid   = tb.t[m, :]

bad_cmp_turns = np.where(~np.all(np.isfinite(cmp_turns_pid), axis=1))[0]
bad_t_turns   = np.where(~np.all(np.isfinite(t_turns_pid), axis=1))[0]

print("plateau_id:", pid)
print("n_turns_in_plateau:", int(np.sum(m)))
print("bad cmp turns (within plateau):", bad_cmp_turns.tolist())
print("bad t turns (within plateau):", bad_t_turns.tolist())

# If there is exactly one bad turn, show how many NaNs it contains
if bad_cmp_turns.size:
    j = bad_cmp_turns[0]
    x = cmp_turns_pid[j, :]
    print("NaNs in cmp of that turn:", int(np.sum(~np.isfinite(x))), "/", x.size)

    # Per-plateau stats
g = df_out.groupby("plateau_id", sort=True)
mean = g.mean(numeric_only=True)
std  = g.std(numeric_only=True)

# Choose x-axis: prefer measured mean current, fallback to I_hint
if "I_mean" in mean.columns:
    x = mean["I_mean"].to_numpy()
elif "plateau_I_hint" in mean.columns:
    x = mean["plateau_I_hint"].to_numpy()
else:
    raise ValueError("No current-like column found for x-axis")

for n in [1, 2, 3]:
    y = mean[f"cmp_A{n}"].to_numpy()
    yerr = std[f"cmp_A{n}"].to_numpy()

    plt.figure()
    # scatter of per-turn values
    plt.plot(df_out["I_mean"].to_numpy(), df_out[f"cmp_A{n}"].to_numpy(), ".", label="per-turn")
    # per-plateau mean with error bars
    plt.errorbar(x, y, yerr=yerr, fmt="o", label="per-plateau mean  std")
    plt.xlabel("current (A)")
    plt.ylabel(f"cmp |C_{n}|")
    plt.title(f"{SEGMENT}: Harmonic amplitude vs current (cmp, n={n})")
    plt.legend()
    plt.show()


```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def phase_reference_from(H_ref, n_ref=1):
    """Return per-turn reference phase phi_ref from harmonic n_ref."""
    orders = H_ref.orders
    j = int(np.where(orders == n_ref)[0][0])
    return np.angle(H_ref.coeff[:, j])

def rotate_harmonics(H, phi_ref):
    """
    Rotate coefficients per turn so that the reference phase is removed.
    C_n_rot = C_n * exp(-i * n * phi_ref)
    """
    orders = H.orders
    C = H.coeff.copy()
    for j, n in enumerate(orders):
        C[:, j] = C[:, j] * np.exp(-1j * n * phi_ref)
    return C

def normal_skew_from_rotated(C_rot, orders, prefix):
    """
    Convert rotated complex coefficients into (Normal, Skew) using:
      N_n = 2*Re(C_n)   for n>=1
      S_n = -2*Im(C_n)  for n>=1
    """
    out = {}
    for j, n in enumerate(orders):
        if int(n) == 0:
            # Optional: keep DC term if you want it
            out[f"{prefix}N0"] = np.real(C_rot[:, j])
            out[f"{prefix}S0"] = 0.0 * np.real(C_rot[:, j])
            continue
        out[f"{prefix}N{int(n)}"] = 2.0 * np.real(C_rot[:, j])
        out[f"{prefix}S{int(n)}"] = -2.0 * np.imag(C_rot[:, j])
    return pd.DataFrame(out)

# 1) Build reference phase from cmp harmonic n=1 (recommended)
phi_ref = phase_reference_from(H_cmp, n_ref=1)

# 2) Rotate both abs and cmp harmonics to the same reference
C_cmp_rot = rotate_harmonics(H_cmp, phi_ref)
C_abs_rot = rotate_harmonics(H_abs, phi_ref)

# 3) Convert to Normal/Skew per turn
df_ns_cmp = normal_skew_from_rotated(C_cmp_rot, H_cmp.orders, prefix="cmp_")
df_ns_abs = normal_skew_from_rotated(C_abs_rot, H_abs.orders, prefix="abs_")

# 4) Attach metadata for grouping (use df_out which already matches kept turns)
meta_cols = [c for c in ["plateau_id", "plateau_step", "plateau_I_hint", "I_mean"] if c in df_out.columns]
df_ns = pd.concat([df_out[meta_cols].reset_index(drop=True), df_ns_abs, df_ns_cmp], axis=1)

# 5) Per-plateau mean  std
if "plateau_id" in df_ns.columns:
    g = df_ns.groupby("plateau_id", sort=True)
    mean_ns = g.mean(numeric_only=True)
    std_ns = g.std(numeric_only=True)
else:
    mean_ns = df_ns.mean(numeric_only=True).to_frame().T
    std_ns = df_ns.std(numeric_only=True).to_frame().T

# 6) Bar chart for one plateau (choose by plateau_id or by nearest current)
PLATEAU_TO_PLOT = 66.0  # change e.g. 0.0, 10.0, 50.0, 66.0, etc.
N_MAX = 20

row = mean_ns.loc[PLATEAU_TO_PLOT]

n = np.arange(1, N_MAX + 1)
N_vals = np.array([row.get(f"cmp_N{k}", np.nan) for k in n])
S_vals = np.array([row.get(f"cmp_S{k}", np.nan) for k in n])

plt.figure()
w = 0.4
plt.bar(n - w/2, N_vals, width=w, label="Normal (cos)")
plt.bar(n + w/2, S_vals, width=w, label="Skew (sin)")
plt.xlabel("harmonic order n")
plt.ylabel("cmp component (phase-referenced)")
plt.title(f"{SEGMENT}: Normal/Skew bars for cmp (plateau_id={PLATEAU_TO_PLOT})")
plt.legend()
plt.show()
```
